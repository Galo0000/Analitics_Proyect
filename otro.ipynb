```python
import pandas as pd
import glob
from IPython.display import display
```

```python
eval_func = lambda x: eval(x)
folder_path = 'Internet'  # Ruta de la carpeta 'Internet'
dataframes = []
# Utilizar glob para encontrar todos los archivos CSV en la carpeta
csv_files = glob.glob(folder_path + '\\*.csv')

# Imprimir los nombres de los archivos CSV encontrados
for file_path in csv_files:
    file_name = file_path.split('\\')[-1]  # Obtener solo el nombre del archivo
    df = pd.read_csv(f'./Internet/{file_name}',delimiter=',')
    df = df.astype(str)
    df = df.applymap(eval_func)
    display(df.dtypes)
    display(df.head())
    dataframes.append()
```
