```python
df = pd.read_csv('./Internet/Acceso_Internet_Fijo_por_rangos_velocidad_bajada_y_provincia.csv',delimiter=',')
df
```

```python
df.isna().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df.dtypes
```

```python
columnas_afectadas = ['HASTA 512 kbps', '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps','+ 6 Mbps - 10 Mbps','+ 10 Mbps - 20 Mbps','+ 20 Mbps - 30 Mbps','+ 30 Mbps','OTROS','Total']
df[columnas_afectadas] = df[columnas_afectadas].astype(str).apply(lambda x: x.str.replace('.', ''))
df
```

```python
df[columnas_afectadas] = df[columnas_afectadas].astype(int)
df
```

```python
# Crear un diccionario de mapeo de nombres de columnas antiguos a nuevos
new_names = {'HASTA 512 kbps': 'hasta 512kb',
              '+ 512 Kbps - 1 Mbps': '512Kb a 1Mb',
                '+ 1 Mbps - 6 Mbps': '1Mb a 6Mb',
                '+ 6 Mbps - 10 Mbps':'6Mb a 10Mb',
                '+ 10 Mbps - 20 Mbps':'10Mb a 20Mb',
                '+ 20 Mbps - 30 Mbps':'20Mb a 30Mb',
                '+ 30 Mbps':'mas de 30Mb',
                'OTROS':'Otros',
                }
# Cambiar los nombres de las columnas utilizando el diccionario
df.columns = [new_names[col] if col in new_names else col for col in df.columns]
df
```

```python
# Utiliza la funcion 'to_pickle' de pandas para guardar en un archivo pickle el dataframe.
df.to_pickle('./Internet_datasets/Acceso_Internet_Fijo_por_rangos_velocidad_bajada_y_provincia.pickle')
```
